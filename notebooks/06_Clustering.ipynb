```python
# Clustering example with K-Means on synthetic blob data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

# Clustering with K-Means
This example builds a simple synthetic dataset, scales it, trains a K-Means model, and visualizes the clusters and centroids. The comments walk through each step so you can reuse the pattern for your own data.

```python
# 1) Create a toy dataset with 4 obvious clusters
X, y_true = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.2,
    random_state=42,
    n_features=2,
)

# 2) Standardize features so K-Means isn't biased by scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3) Fit K-Means
#    n_init controls how many random initializations we try
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# 4) Recover the centroids in original (unscaled) feature space
centroids_scaled = kmeans.cluster_centers_
centroids = scaler.inverse_transform(centroids_scaled)

# 5) Visualize points and centroids
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="tab10", s=30, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c="black", s=180, marker="X", label="Centroids")
plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.tight_layout()
plt.show()

# 6) Quick numeric check: cluster sizes
unique, counts = np.unique(labels, return_counts=True)
dict(zip(unique, counts))
```
