## Load the dataset and import preprocessing tools

This first section imports the scikit-learn helpers used throughout the notebook and loads the Iris dataset. `load_iris` returns the feature matrix `x` and target labels `y`.

The goal here is simply to get a small, clean dataset that makes preprocessing steps easy to demonstrate.

## What this notebook demonstrates

This notebook walks through the most common preprocessing steps in scikit-learn:

- splitting data into training and test sets
- scaling numeric features with standardization and min-max normalization
- encoding categorical features with ordinal encoding and one-hot encoding

Each block below shows the raw code first, then the result of applying the preprocessing step.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Split the data into training and test sets

`train_test_split` separates the dataset into two parts: one for training the preprocessing step and model, and one for testing how well the processing generalizes to unseen data.

Using `random_state=42` makes the split reproducible, so the same rows go into the same sets each time the notebook is run.

```python
x, y = load_iris(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)   
```

```python
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

## Standardize the numeric features

`StandardScaler` transforms each feature so it has mean 0 and standard deviation 1. This is useful when features are on different scales because many machine learning algorithms work better when the inputs are centered and scaled.

The scaler is fit only on `x_train`, then reused to transform `x_test`. That avoids leaking information from the test set into the training process.

```python
import numpy as np
(x_train-np.mean(x_train, axis=0)) / np.std(x_train, axis=0  )
```

## Check the scaling math manually

This cell recomputes standardization with NumPy so you can see what `StandardScaler` is doing under the hood.

The formula is:

$\displaystyle z = \frac{x - \mu}{\sigma}$

where `x` is the original value, `μ` is the feature mean, and `σ` is the feature standard deviation.

```python

x_min = np.min(x_train, axis=0)
x_max = np.max(x_train, axis=0)
x_train_scaled = (x_train - x_min) / (x_max - x_min)
x_test_scaled = (x_test - x_min) / (x_max - x_min) 
```

## Apply min-max normalization

This section rescales every feature into the range from 0 to 1.

The formula is:

$\displaystyle x' = \frac{x - x_{min}}{x_{max} - x_{min}}$

This is different from standardization: instead of centering around a mean of 0, it compresses values into a fixed interval based on the minimum and maximum values seen in the training data.

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

## Encode categorical features

The next examples use the `car` and `adult` datasets from OpenML to show how categorical variables are transformed into numeric form.

First, `OrdinalEncoder` is used when the categories have a natural order. Then `OneHotEncoder` is used when the categories are nominal and should not be treated as ordered values.

```python
data = fetch_openml('car', as_frame=True)
data.frame.head()
```

```python
columns_to_encode = ['lug_boot', 'safety']
encoder = OrdinalEncoder(
    categories=[
        ['small', 'med', 'big'],
        ['low', 'med', 'high'],
    ]
)
data.frame[columns_to_encode] = encoder.fit_transform(data.frame[columns_to_encode])
```

```python
data = fetch_openml('adult', as_frame=True)
data.frame.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data.occupation.value_counts()
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_values = encoder.fit_transform(data.frame[['occupation', 'education']])
new_cols = encoder.get_feature_names_out(['occupation', 'education'])
encoded_df = pd.DataFrame(encoded_values, columns=new_cols)
```

```python
new_cols
```

```python
df_encoded = pd.concat([data.frame, encoded_df], axis=1)
df_encoded.drop(columns=['occupation', 'education'], inplace=True)  

data_final = pd.concat([data.frame, encoded_df], axis=1)
data_final.drop(columns=['occupation', 'education'], inplace=True)  
```
