# Splitting data



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
x, y = load_iris(return_X_y=True)   
```

```python
x, y = load_iris(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,   test_size=0.2, random_state=42)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
counts = np.bincount(y_train)
positions = np.arange(3)
plt.bar(positions, counts)

plt.xticks(positions, counts)
plt.xticks(positions, data.target_names)
```

```python
from sklearn.model_selection import StratifiedGroupKFold

split = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=42)
for train_idx, test_idx in split.split(x, y, groups=y):
    print(f"Train: {train_idx}, Test: {test_idx}")  
```
